```python
# Cell 3: DataFrame
import pandas as pd

df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "Age": [28, 34, 45, 31, 22],
    "City": ["New York", "London", "Tokyo", "Paris", "Berlin"],
    "Salary": [75000, 82000, 95000, 68000, 55000],
    "Active": [True, False, True, True, False]
})
df
```

```python
# Cell 4: DataFrame describe + info
df.describe()
```

```python
# Cell 5: Matplotlib plot
import matplotlib.pyplot as plt
import numpy as np

print("Plotting trig functions...")
print("This should show a plot of sin and cos functions.")
print("If you see the plot, then Matplotlib is working correctly.")
x = np.linspace(0, 10, 100)
plt.figure(figsize=(8, 4))
plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)")
plt.title("Trig Functions")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Cell 6: Bar chart from DataFrame
df.plot.bar(x="Name", y="Salary", figsize=(8, 4), color="teal")
plt.title("Salary by Person")
plt.ylabel("Salary ($)")
plt.tight_layout()
plt.show()
```

```python
# Cell 7: Scatter plot
print("Plotting random scatter...")
print("This should show a scatter plot with random colors and sizes.")
print("If you see the scatter plot, then Matplotlib is working correctly.")
plt.figure(figsize=(6, 6))
np.random.seed(42)
plt.scatter(
    np.random.randn(200),
    np.random.randn(200),
    c=np.random.rand(200),
    s=np.random.rand(200) * 300,
    alpha=0.5,
    cmap="viridis"
)
plt.colorbar()
plt.title("Random Scatter")
plt.show()
```
